```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
print("Beginning Data Retrieval")
print("------------------------")
cityweatherlist = []

base_url = "https://api.openweathermap.org/data/2.5/weather?units=imperial"
base_url = base_url + "&appid=" + weather_api_key
for city in cities:
    print("retrieving city: %s", city)
    url = base_url + "&q=" + city
    
    try: 
        cityweather = requests.get(url)

        
        # appid={your api key}
        cityweather = requests.get(url).json()


#         city = 'cape town'
        lat = cityweather["coord"]["lat"]
        lng = cityweather["coord"]["lon"]
        maxtemp = cityweather["main"]["temp_max"]
        windspeed = cityweather["wind"]["speed"]
        humidity = cityweather["main"]["humidity"]
        date = cityweather["dt"]
        country = cityweather["sys"]["country"]
        cloudiness = cityweather["clouds"]["all"]
        weatherobject = {'City':city,
                         'Cloudiness':cloudiness,
                        'Country':country,
                        'Date':date,
                        'Humidity':humidity,
                        'Lat':lat,
                        'Lng':lng,
                        'Max Temp':maxtemp,
                        'Wind Speed':windspeed}

        cityweatherlist.append(weatherobject) 
    except: 
        print(cityweather)
        print(url)
        print("city not found skipping")
        pass
print("completed")
cityweatherlist
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_hem_df = cities_df.loc[pd.to_numeric(cities_df["Lat"]).astype(float) > 0]
southern_hem_df = cities_df.loc[pd.to_numeric(cities_df["Lat"]).astype(float) < 0]

# north_hem_df.head()
# southern_hem_df.head(10)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hem_df['Lat']
y_values = north_hem_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
print(f"The Regression Line Equation is: {line_eq}")
print(f"The r-squared is: {rvalue}")
plt.savefig("output_data/northHem_lat_vs_maxTempRegression.png")
plt.show()
```

Max temp and latitude have a strong correlation in the Northern Hemisphere.
